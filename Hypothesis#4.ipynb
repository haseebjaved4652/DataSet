```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the CSV file into a pandas DataFrame
data = pd.read_csv("Cab_Data.csv")

# Assuming there is a "Date of Travel" column with numeric values
data['Date of Travel'] = pd.to_datetime('1899-12-30') + pd.to_timedelta(data['Date of Travel'], unit='D')

# Filter data for February of 2016, 2017, and 2018
filtered_data = data[(data['Date of Travel'].dt.year.isin([2016, 2017, 2018])) & (data['Date of Travel'].dt.month == 2)]

# Grouping the filtered data by the "City" column and count the number of transactions in each city
city_transaction_counts = filtered_data.groupby("City")["Transaction ID"].count()

# Set Seaborn style
sns.set(style="whitegrid")

# Creating a bar graph
plt.figure(figsize=(12, 6))
colors = sns.color_palette("pastel")
city_transaction_counts.plot(kind="bar", color=colors)
plt.title("Number of Transactions per City (February 2016, 2017, 2018)", fontsize=16)
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Transactions", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# Display the graph
plt.show()
```
