```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the CSV file into a pandas DataFrame
data = pd.read_csv("Cab_Data.csv")

# Grouping the data by the "City" column and count the number of transactions in each city
city_transaction_counts = data["Transaction ID"].groupby(data["City"]).count()

# Creating a bar graph
plt.figure(figsize=(10, 6))
city_transaction_counts.plot(kind="bar")
plt.title("Number of Transactions per City")
plt.xlabel("City")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=45)
plt.tight_layout()

# A diplay of  the graph
plt.show()
```
