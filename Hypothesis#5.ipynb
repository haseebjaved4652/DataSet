```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file into a DataFrame
data = pd.read_csv("City.csv")

# Extracting the "City," "Population," and "Users" columns
cities = data["City"]
# Removing commas and convert "Population" and "Users" columns to numeric (integer or float)
population = pd.to_numeric(data["Population"].str.replace(',', ''))
users = pd.to_numeric(data["Users"].str.replace(',', ''))

# Calculating the user-to-population ratio for each city
user_population_ratio = users / population

# Defining colors for the bars
colors = ['skyblue', 'lightgreen', 'lightcoral', 'lightsalmon', 'lightseagreen']

# Creating a bar graph for the user-to-population ratio with custom colors
plt.figure(figsize=(12, 6))  # Adjust the figure size if needed
bars = plt.bar(cities, user_population_ratio, label="User-to-Population Ratio", color=colors)

# Adding labels and title
plt.xlabel("Cities")
plt.ylabel("Ratio")
plt.title("User-to-Population Ratio in Different Cities")

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Adding the number of users with commas on each bar
for bar, user_count in zip(bars, users):
    formatted_user_count = "{:,}".format(user_count)  # Format the number with commas
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), formatted_user_count,
             ha='center', va='bottom')

# A diplay of  the graph
plt.tight_layout()
plt.show()
```
