```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading CSV file into a DataFrame
df = pd.read_csv('Cab_Data.csv')

# Grouping the data by Company and sum the Cost of Trip
company_cost = df.groupby("Company")["Cost of Trip"].sum()

# Plotting the bar graph
plt.figure(figsize=(10, 6))
colors = ["pink", "yellow"]
company_cost.plot(kind="bar", color=colors, alpha=0.7)
plt.title("Accumulated Cost of Trip by Company")
plt.xlabel("Company")
plt.ylabel("Accumulated Cost of Trip")
plt.xticks(rotation=0)

# Show the plot
plt.tight_layout()
plt.show()
```
